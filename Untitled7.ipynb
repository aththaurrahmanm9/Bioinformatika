```python
import numpy as np
import matplotlib.pyplot as plt

labels = ["Ved", "Vsc", "Vtr", "Vsa", "Vcs", "Vti"]
n_vars = len(labels)

data = {
    "Kawengan Anticline": [3.67, 4.00, 3.00, 3.00, 3.00, 3.00],
    "Wonocolo Formation": [3.33, 3.67, 3.00, 3.00, 3.00, 2.00],
    "Kayangan Api":       [3.00, 3.00, 4.00, 3.67, 3.00, 4.00]
}

angles = np.linspace(0, 2 * np.pi, n_vars, endpoint=False)
angles_closed = np.concatenate((angles, [angles[0]]))

fig, ax = plt.subplots(subplot_kw={"polar": True}, figsize=(6, 6))

for name, values in data.items():
    vals = values + [values[0]]
    ax.plot(angles_closed, vals, linewidth=1.8, label=name)
    ax.fill(angles_closed, vals, alpha=0.15)

ax.set_xticks(angles)
ax.set_xticklabels(labels)

ax.set_ylim(0, 4)
ax.set_yticks([1, 2, 3, 4])
ax.set_yticklabels(["1", "2", "3", "4"])

ax.set_title("Radar diagram of key geosites by evaluation dimension", pad=18)
ax.legend(loc="upper right", bbox_to_anchor=(1.45, 1.05))

plt.tight_layout()
plt.show()
```
